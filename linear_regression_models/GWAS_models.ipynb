```r
#############################
####  Linear  Modelling  ####
#############################

#################
## SNP1  MODEL ##
#################

# Making a linear model that accounts for the ancestry as a factor
model_snp1 <- lm(risk_score ~ SNP1 + ancestry, data = gwas_data)
summary(model_snp1) 

# Checking the linear model assumptions 
plot(model_snp1)

# Making interaction model to check for the ancestry influence
snp1_interaction <- lm(risk_score ~ SNP1 * ancestry, data = gwas_data)
summary(snp1_interaction)

#################
## SNP2  MODEL ##
#################

# Making a linear model that accounts for the ancestry as a factor
model_snp2 <- lm(risk_score ~ SNP2 + ancestry, data = gwas_data)
summary(model_snp2) 

# Checking the linear model assumptions 
plot(model_snp2)

# Making interaction model to check for the ancestry influence
snp2_interaction <- lm(risk_score ~ SNP2 * ancestry, data = gwas_data)
summary(snp2_interaction)

#################
## SNP3  MODEL ##
#################

# Making a linear model that accounts for the ancestry as a factor
model_snp3 <- lm(risk_score ~ SNP3 + ancestry, data = gwas_data)
summary(model_snp3) 

# Checking the linear model assumptions 
plot(model_snp3)

# Making interaction model to check for the ancestry influence
snp3_interaction <- lm(risk_score ~ SNP3 * ancestry, data = gwas_data)
summary(snp3_interaction)


#################
## SNP4  MODEL ##
#################

# Making a linear model that accounts for the ancestry as a factor
model_snp4 <- lm(risk_score ~ SNP4 + ancestry, data = gwas_data)
summary(model_snp4) 

# Checking the linear model assumptions 
plot(model_snp4)

# Making interaction model to check for the ancestry influence
snp4_interaction <- lm(risk_score ~ SNP4 * ancestry, data = gwas_data)
summary(snp4_interaction)


#################
## SNP5  MODEL ##
#################

# Making a linear model that accounts for the ancestry as a factor
model_snp5 <- lm(risk_score ~ SNP5 + ancestry, data = gwas_data)
summary(model_snp5) 

# Checking the linear model assumptions 
plot(model_snp5)

# Making interaction model to check for the ancestry influence
snp5_interaction <- lm(risk_score ~ SNP5 * ancestry, data = gwas_data)
summary(snp5_interaction)
```
